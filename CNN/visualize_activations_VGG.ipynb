### Visualize Activations

```python
from matplotlib import pyplot as plt
%matplotlib inline

from keras import activations
from keras.applications import VGG16
from vis.utils import utils
from vis.visualization import visualize_activation
```

```python
# Build the VGG16 network with ImageNet weights
model = VGG16(weights='imagenet', include_top=True)

# Utility to search for layer index by name (the last layer)
layer_idx = utils.find_layer_idx(model, 'predictions')

# Swap softmax with linear (for visualizing activations)
model.layers[layer_idx].activation = activations.linear
model = utils.apply_modifications(model)
```

```python
# Define figure size
plt.rcParams['figure.figsize'] = (18, 6)

# 288 is index for leopard
img = visualize_activation(model, layer_idx, filter_indices=288)
plt.imshow(img)
```

#### Converge the model ont the output class
What does the net thinks is a leopard?

```python
img = visualize_activation(model, layer_idx, filter_indices=288, max_iter=500, verbose=True)
```

```python
plt.imshow(img)
```

#### Hallucinations
By passing a list of values to *filter_indices*, the model tries to combines the activations of both classes 

```python
# 20 is the index for ouzel
img = visualize_activation(model, layer_idx, filter_indices=[288,20], max_iter=500)
plt.imshow(img)
```
